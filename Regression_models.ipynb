#2. What are the key assumptions of Simple Linear Regression?

The key assumptions of Simple Linear Regression (SLR) ensure that the model is valid, reliable, and interpretable. These assumptions relate to the relationship between the independent variable (X) and the dependent variable (Y), and to the behavior of the error term.

Key Assumptions of Simple Linear Regression:-

Linearity:

The relationship between X and Y is linear.

The mean of Y changes linearly with X.

Violation: curved or non-linear patterns in the data.

Independence of Errors:

The residuals (errors) are independent of each other.

Commonly violated in time-series data.

Checked using the Durbinâ€“Watson test.

Homoscedasticity (Constant Variance):

The variance of residuals is constant across all values of X.

Violation (heteroscedasticity): residual spread increases or decreases with X.

Normality of Errors:

The residuals are normally distributed with a mean of zero.

Important for valid hypothesis testing and confidence intervals.

Checked using Qâ€“Q plots or statistical tests.

No Perfect Multicollinearity:

In SLR, this means X should not be constant (there must be variability in X).

Though multicollinearity is mainly a concern in multiple regression, SLR still requires X to vary.
